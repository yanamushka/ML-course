1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import sympy as sp
x = sp.symbols('x')
f_x = 10 * x - 5
derr_f = sp.diff(f_x, x)

```

```python
import matplotlib.pyplot as plt
import numpy as np
x1  = np.linspace(0,10)
f = 10 * x1 - 5
df = 10 + x1 * 0
plt.plot(x1, f, label = 'f(x) = 10x - 5', color = 'blue')
plt.plot(x1, df, label = "f'(x) = 10", color = 'red')
plt.grid()
plt.legend()
```

З кожною додатковою годиною навчання оцінка збільшується на 10 одиниць.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python

sin_x = sp.sin(x)
df_sin_x1 = sp.diff(sin_x, x)

sin_x1 = np.sin(x1)
df_sin_x1 = np.cos(x1)

plt.plot(x1, sin_x1, label = 'f(x) = sin_x', color = 'blue')
plt.plot(x1, df_sin_x1, label = "f'(x) = cos(x)", color = 'red')
plt.legend()
```

```python
ln_x = sp.ln(x)
df_ln_x = sp.diff(ln_x, x)

ln_x1 = np.log(x1)
df_ln_x1 = 1 / x1

plt.plot(x1, ln_x1, label = 'f(x) = ln(x)', color = 'blue')
plt.plot(x1, df_ln_x1, label = "f'(x) = 1/x", color = 'red')
plt.legend()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
f_spending = x**2 - 5 * x + 1
df_f_spending = sp.diff(f_spending, x)

x2 = np.linspace(0, 100)
f_spending_x2 = x2**2 - 5 * x2 + 1
df_f_spending_x2 = 2 * x2 - 5
plt.plot(x2, f_spending_x2, label = 'f(x) = x2**2 - 5 * x2 + 1', color = 'blue')
plt.plot(x2, df_f_spending_x2, label = "f'(x) = 2 * x2 - 5", color = 'red')
plt.legend()
plt.grid()
```
